##KNN
* K Nearest Neighbours
* Uses similarity measure ( how much two objects are alike)
* Stores available cases and for finding label for new case, check in it's neighbor, and says that I am one of them.
* Applications
    1. Recommended System (E Commerce Websites)
    2. Concept Search (Internet generates plethora of documents each day, to segeregate them we may use this algorithm) 

#### My Implementation

* Algorithm
    1. Find the distance between new_data_instance and all existing instances
    2. Find nearest K points.
    3. Among them choose which target class occurred in majority.

* Catch
  1. Choose approaprite value for K

* Steps
  1. Define Distance Function (Euclidian)
  2. Define FindNeighter Function
  3. Define Predict Function
  4. Define Evaluation Function

* input_dataset: input_dataset on from which we need to build algorithm
* output_dataset: corresponding output/target of input_dataset
* new_instance: single instance (row) for which we need to find the target
* input_test_dataset: whole dataset for which we need to find target
* output_test_dataset: output of the test dataset (we use this for evaluation)
* K: K neighbors




````python
def EuclidianDistance(input_dataset,new_instance):
  distance=[]
  for old_instance in input_dataset:
    dis=np.sqrt(np.sum(np.power(old_instance-new_instance,2)))
    distance.append(dis)
  return np.array(distance)
````

````python
def FindNeighbor(K,distance,output_dataset):
  #concate=pd.concat([distance,output_dataset],axis=1)
  #cannot concatenate object of type '<class 'numpy.ndarray'>'; only Series and DataFrame objs are valid
  concate=pd.concat([pd.Series(distance),pd.Series(output_dataset)],axis=1)
  concate=concate.sort_values(0)
  neighbors=concate[:K][1]
  return np.array(neighbors)
````

````python
def predict(neighbors):
  #[0,1,1] #1
  return np.bincount(neighbors).argmax()
  
````

It is coming correct for single test case

````python
sample_input=[10,22,22,12]
neighbors=FindNeighbor(3,EuclidianDistance(input_train,sample_input),output_train)
p=predict(neighbors)
print(p)
print(dataset['target_names'][p])
````

````python
def prediction(input_dataset,output_dataset,input_test_dataset,K):
  predicted=[]
  for i in input_test_dataset:
    neighbors=FindNeighbor(K,EuclidianDistance(input_dataset,i),output_dataset)
    predicted.append(predict(neighbors))
  return np.array(predicted)
````

````python
def Score(predicted,output_test_dataset):
  return np.mean(predicted == output_test_dataset)
````

````python
output_test_predicted=prediction(input_train,output_train,input_test,3)
````

````python
score=Score(output_test_predicted,output_test)
print("predicted output:",output_test_predicted)
print("actual output:",output_test)
print("Score %.3f"%(score))
````
