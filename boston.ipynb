```
from sklearn.model_selection import train_test_split
input_train,input_test,output_train,output_test=train_test_split(boston.data,boston.target,random_state=42)
```

```
#lets make histogram
from matplotlib import pyplot

for i in range(13):
  print(boston.feature_names[i])
  pyplot.hist(input_train[:,i],label=boston.feature_names[i])
  pyplot.show()




```

```
#let's see correlation between all 13 features and targer feature
for i in range(13):
  pyplot.plot(output_train,input_train[:,i])
  pyplot.ylabel(boston.feature_names[i])
  pyplot.xlabel("Price in $1000")
  pyplot.show()
```

```
#lets see range
for i in range(i):
  print(boston.feature_names[i],end=" ")
  print(input_train[:,i].min(),end=" ")
  print(input_train[:,i].max())
```

```
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(input_train,output_train)
print("Training data Accuracy:%.3f"%lr.score(input_train,output_train))
print("Testing data Accuracy:%.3f"%lr.score(input_test,output_test))
```

```
output_test_from_linear=lr.predict(input_test)
for i in range(10):
  print(output_test[i],"%.3f"%output_test_from_linear[i])
```

```
from sklearn.neighbors import KNeighborsRegressor
for i in range(3,10):
  knn=KNeighborsRegressor(n_neighbors=i)
  knn.fit(input_train,output_train)
  print("K=",i)
  print("Training data Accuracy:%.3f"%knn.score(input_train,output_train))
  print("Testing data Accuracy:%.3f"%knn.score(input_test,output_test))
```

```
from sklearn.linear_model import Ridge
for i in [1.0,10,0.1]:
  ridge=Ridge(alpha=i)
  ridge.fit(input_train,output_train)
  print("alpha=",i)
  print("Training data Accuracy:%.3f"%ridge.score(input_train,output_train))
  print("Testing data Accuracy:%.3f"%ridge.score(input_test,output_test))

  
```

```
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor()
tree.fit(input_train,output_train)
print("Training data Accuracy:%.3f"%tree.score(input_train,output_train))
print("Testing data Accuracy:%.3f"%tree.score(input_test,output_test))
```

Here we observed overfitting. 
Decision Tree didn't work well with unknown data

```
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators=7,random_state=4)
forest.fit(input_train,output_train)
print("Training data Accuracy:%.3f"%forest.score(input_train,output_train))
print("Testing data Accuracy:%.3f"%forest.score(input_test,output_test))
```

Model forest is the good model by far for predicting median value of price of occupied homes. :)

```
#requires pre processing
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(input_train)
input_train_scaled=scaler.transform(input_train)
input_test_scaled=scaler.transform(input_test)
from sklearn.svm import SVR
for c in [0.001, 0.01, 0.1, 1, 10, 100]:
  for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    svr=SVR(C=c,gamma=gamma)
    print("C:",c,"gamma",gamma)
    svr.fit(input_train_scaled,output_train)
    print("Training data Accuracy:%.3f"%svr.score(input_train_scaled,output_train))
    print("Testing data Accuracy:%.3f"%svr.score(input_test_scaled,output_test))

```
